```python
#importing wanted libraries

import pandas as pd
import numpy as np
import requests
import cv2
from sklearn.utils import shuffle

from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
#creating dataframe for george and non_george

georges_df = pd.read_csv('/content/georges.csv',header=None)
georges_df.columns=['img_path']
non_georges_df = pd.read_csv('/content/non_georges.csv',header=None)
non_georges_df.columns=['img_path']
```

```python
#adding a label column for george and non_george

georges_df['label'] = 1
non_georges_df['label'] = 0
```

```python
print("Georges DataFrame:")
print(georges_df.head())

print("\nNon-Georges DataFrame:")
print(non_georges_df.head())
```

```python
#combining george dataframe and non_george dataframe

combined_df = pd.concat([georges_df, non_georges_df], ignore_index=True)
```

```python
print(f"Georges DataFrame has {georges_df.shape[0]} rows")
print(f"Non-Georges DataFrame has {non_georges_df.shape[0]} rows")
print(f"Combined DataFrame has {combined_df.shape[0]} rows")
```

```python
#where we use shuffle to print label without an order,thus time we get an idea about 0 and 1 present or not

combined_df = shuffle(combined_df, random_state=42)
combined_df = combined_df.reset_index(drop=True)
```

```python
print(combined_df.head(10))
```

```python
print(combined_df.head())
print(combined_df.columns)
```

```python
#load images from url

def load_images_from_urls(df):
    images = []
    labels = []
    for index, row in df.iterrows():
        response = requests.get(row['img_path'])
        img_array = np.asarray(bytearray(response.content), dtype=np.uint8)
        img = cv2.imdecode(img_array, cv2.IMREAD_COLOR)
        img = cv2.resize(img, (224, 224))
        images.append(img)
        labels.append(row['label'])
    return np.array(images), np.array(labels)
```

```python
#input data will collect on images, and target data will collect on labels

images,labels=load_images_from_urls(combined_df)
```

```python
#model training

train_images,test_images,train_labels,test_labels=train_test_split(images,labels,test_size=0.2,random_state=42)


#normalization

train_images=train_images/255.0
test_images=test_images/255.0

```

```python
#CNN architecture

import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout, Input

# This is my CNN model
model = Sequential([
    Input(shape=(224, 224, 3)),
    Conv2D(32, (3, 3), activation='relu'),
    MaxPooling2D((2, 2)),
    Conv2D(64, (3, 3), activation='relu'),
    MaxPooling2D((2, 2)),
    Conv2D(128, (3, 3), activation='relu'),
    MaxPooling2D((2, 2)),
    Flatten(),
    Dense(512, activation='relu'),
    Dropout(0.5),
    Dense(1, activation='sigmoid')  # Binary classification
])

model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

model.fit(train_images, train_labels, epochs=10, validation_data=(test_images, test_labels))

loss, accuracy = model.evaluate(test_images, test_labels)
print(f'Test accuracy: {accuracy:.2f}')
```

```python
#predict using test image,not test data image,its only for printing accuracy score etc

predictions = model.predict(test_images)
predictions = (predictions > 0.5).astype(int)

# Generate classification report
print(classification_report(test_labels, predictions))
```

```python
#test image loading

import cv2
import numpy as np

def load_and_preprocess_image(image_path):
    img = cv2.imread(image_path)
    img = cv2.resize(img, (224, 224))
    img = np.array(img) / 255.0
    img = np.expand_dims(img, axis=0)

    return img
```

```python
#test image predicting

def predict_george_presence(model, image_path):
    processed_image = load_and_preprocess_image(image_path)
    prediction = model.predict(processed_image)
    if prediction[0] > 0.5:
        return "George is present in the image."
    else:
        return "George is not present in the image."
```
